### Сглаживание средним окна, окно в 55, + лин. регрессия

```python
import keras
import numpy as np
import scipy as sp
import scipy.stats

x_train = np.load('data/x_train.npy')
y_train = np.load('data/y_train.npy')
x_test = np.load('data/x_test.npy')
y_test = np.load('data/y_test.npy')

def get_model(N):
    model = keras.models.load_model(f'models/model_aenc_{N}')
    return model

colors=[color_11, color_12, color_13, color_14, 'black']
i=0;

fig, axes = plt.subplots(5,2, figsize=(15,12))


for N in [50, 100, 250, 500, 1000]:
    model = get_model(N)
    x_arr = np.mean(x_train, axis=0) 
    pred = np.mean(np.array(model.call(x_train)), axis=0)
    ls='dashed'
    ax1 = axes[i, 1]
    ax1.title.set_text(f'AMSE against R,  {N} neurons')
    delta = (pred-x_arr)*(pred-x_arr)
    delta_flatten = np.convolve(delta, np.ones((55,))/55, mode='valid')
    ax1.plot(delta_flatten, color=colors[i], label=f'{N} neurons')
    i+=1
    
    
colors=[color_11, color_12, color_13, color_14, 'black']
i=0;
for N in [50, 100, 250, 500, 1000]:
    model = get_model(N)
    x_arr = np.mean(x_train, axis=0) 
    pred = np.mean(np.array(model.call(x_train)), axis=0)
    ls='dashed'
    ax1 = axes[i, 0]
    ax1.title.set_text(f'RMSE against R,  {N} neurons')
    delta = (pred-x_arr)/x_arr*(pred-x_arr)/x_arr
    delta_flatten = np.convolve(delta, np.ones((55,))/55, mode='valid')
    slope, intercept, r_value, p_value, std_err = sp.stats.linregress(x=[i for i in range(0,4630)][50:], y=delta[50:])
    ax1.plot(delta_flatten, color=colors[i], label=f'{N} neurons')
    x = np.array([i for i in range(0,4576)])
    ax1.plot(x,slope*x+intercept, linestyle='dashed', color=colors[i], label=f'{N} neurons')
    i+=1    

plt.tight_layout()
plt.show()
```
